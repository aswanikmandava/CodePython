# Pie Chart

```python
from matplotlib import pyplot

mydata = [item for item in range(0, 100, 10)]
pyplot.pie(mydata)
```

# Line Chart

```python
x_axis = [1, 2, 3, 4]
y_axis = [10, 40, 50, 30]
pyplot.plot(x_axis, y_axis)
```

# Bar Chart

```python
x_axis = [item for item in range(6)]
y_axis = [55, 43, 20, 78, 13, 20]
pyplot.bar(x_axis, y_axis)
```

# Scatter Chart

```python
pyplot.scatter(x_axis, y_axis)
```

# Bubble Chart
This chart extends the concept of a scatter plot by adding a third dimension to the data points. In addition to the x-axis and y-axis, the size of the bubbles represents a third variable

```python
x_axis = [item for item in range(6)]
y_axis = [55, 43, 20, 78, 13, 20]
weight = [450, 5000, 2000, 600, 400, 700]
pyplot.scatter(x_axis, y_axis, weight)
```
